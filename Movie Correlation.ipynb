```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in data
df = pd.read_csv(r'C:\Users\Kelvin\Documents\GitHub\DataAnalyticsProjects\movies.csv')
```

```python
# Look at first 5 entries
df.head()
```

```python
# CLEANING/MUNGING DATA
```

```python
# Loop through the data to see if there's any missing data
for column in df.columns:
    percent_missing = np.mean(df[column].isnull())
    print('{} - {}%'.format(column, round(percent_missing*100)))
```

```python
# Check column data types
print(df.dtypes)
```

```python
# Fix NaNs that are casted as float to be 0 so they can be changed to int
df['budget'] = df['budget'].fillna(0).astype(np.int64)

# Change data types of columns (formatting)
df['budget'] = df['budget'].astype(np.int64)

# Do the same for gross
df['gross'] = df['gross'].fillna(0).astype(np.int64)
df['gross'] = df['gross'].astype(np.int64)

df.head()
```

```python
# Drop duplicates rows
df.drop_duplicates()
```

```python
# Check for outliers
df.boxplot(column=['gross'])
```

```python
# Sort data by gross (descending)
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# FINDING CORRELATION
```

```python
# Hypothesis: An increase in movie budget is correlated to an increase in gross earnings.

# Scatter plot, budget vs. gross earnings
plt.scatter(x=df['budget'], y=df['gross'])

# Scatter plot formatting
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')

plt.show()
```

```python
# Use seaborn to plot budget vs. gross and linear regression model fit
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Correlation matrix between all numeric columns
df.corr(method='pearson')
```

```python
# High correlation between budget and gross (0.75)
```

```python
# Use seaborn to visualize the correlation matrix with a heatmap
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
# Numerize non-numeric values to be used in correlation

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Use seaborn to visualize the correlation matrix with a heatmap
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

print(high_corr)
```
